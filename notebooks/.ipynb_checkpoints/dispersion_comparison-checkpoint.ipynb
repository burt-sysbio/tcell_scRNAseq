data sets:

methods:
- scanpy default normalization
- hafemeister/satija (seurat authors) new norm. methods (neg. binomial distr.)

```python
import scanpy as sc #for scanpy >= 1.3.7
import anndata as ann
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
from matplotlib import colors
sc.settings.verbosity = 3             # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.logging.print_versions()
sc.settings.set_figure_params(dpi=80, facecolor='white')

import seaborn as sns
```

```python
#path = "/home/burt/Documents/projects/2020/tcell_scRNAseq/output/"
path = "C:/Users/Philipp/OneDrive/projects/2020/tcell_scRNAseq/output/"

# create list of filenames, for seurat add _seurat
filenames_sc = ["variance_xin_2018", "variance_ciucci_2019", "variance_magen_2019"]
filenames_seu = [name+"_seurat" for name in filenames_sc]

# append path and csv to read dataframes
filenames_sc = [path+name+".csv" for name in filenames_sc]
filenames_seu = [path+name+".csv" for name in filenames_seu]

# add data sets that are specific to seurat
seu_only = ["variance_pbmc3k_10x_seurat", "variance_pbmc3k_10x_tcells_seurat"]
seu_only = [path+name+".csv" for name in seu_only]
filenames_seu = filenames_seu + seu_only

df_sc = [pd.read_csv(name) for name in filenames_sc]
df_seu = [pd.read_csv(name) for name in filenames_seu]

# get study names w.o. variance
studies_sc = ["xin2018", "ciucci2019", "magen2019"]
studies_seu = ["xin2018", "ciucci2019", "magen2019", "10x pbmc", "10x T cells"]

# add normalization method and study name to each data frame
for df, study in zip(df_sc, studies_sc):
    df["study"] = study
    df["norm."] = "scanpy"

for df, study in zip(df_seu, studies_seu):
    df["study"] = study
    df["norm."] = "sctransform"

df = pd.concat(df_sc)
df_seu = pd.concat(df_seu)
```

# normalized gene variance distribution using scanpy normalization

```python
df2 = df[df.highly_variable == True]
df2["dispersion_log"] = np.log2(df2.dispersions_norm)

g = sns.catplot(data = df, x = "study", y = "dispersions_norm", kind = "violin")
g = sns.relplot(data = df, x = "means", y = "dispersions_norm", col = "study")
```

# normalization using negative binomial distribution normalization method (Hafemeister/Satija 2019)
one can see that mean/var trend is eliminated but normalized variance distribution look very similar

```python
df_seu["log_gmean"] = np.log2(df_seu["sct.gmean"])
df_seu["log_res_var"] = np.log2(df_seu["sct.residual_variance"])
g = sns.catplot(data = df_seu, x = "study", y = "log_res_var", kind = "violin", aspect = 1.5)
g.set(ylabel = "log(gene variance norm.)")
# clip outliers!
g.savefig("scseq_variance_distribution.pdf")
g = sns.relplot(data = df_seu, x = "log_gmean", y = "log_res_var", col = "study")

g.savefig("scseq_mean_variance_relation.pdf")
```
